```python
from scipy.stats import linregress
import matplotlib.pyplot as plt
```

```python
temps = [2,5,6,0,9]
daylight = [6,8,9,5,5]
```

```python
plt.scatter(temps, daylight)
```

```python
linregress(daylight,temps)
```

```python
(slope,intercept,rval,pval,stderr) = linregress(daylight,temps)
```

```python
lineEq = "y ="+ str(round(slope,2)) + "x +" + str(round(intercept,2))
lineEq
```

```python
regressV = [(day * slope + intercept) for day in daylight]
```

```python
plt.scatter(daylight, temps)

plt.plot(daylight, regressV)
```

```python
# Create an equal number of latitudes and temperatures.
lats = [42.5, 43.9, 8.1, 36.8, 79.9, 69.1, 25.7, 15.3, 12.7, 64.5]
temps = [80.5, 75.3, 90.9, 90.0, 40.4, 62.3, 85.4, 79.6, 72.5, 72.0]
```

```python
(slope, intercept, rValue, pValue, stdError) = linregress(lats, temps)

lineEquation = "y = " + str(round(slope,2))  + "x + " + str(round(intercept,2))

print(lineEquation)
print(f"the pvalue is: {pValue:.3f}")
```

```python
regressValues = [(lat * slope + intercept) for lat in lats]
```

```python
plt.scatter(lats,temps)

plt.plot(lats, regressValues, "r")\


plt.annotate(lineEquation, (10,40), fontsize=15, color="red")
plt.xlabel = ("Latitude")
plt.ylabel = ("Temp")
plt.show()
```
